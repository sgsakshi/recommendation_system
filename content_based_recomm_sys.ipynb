## Logic Building

```python
movies.iloc[0,3]                             ## genere
```

```python
# we want the keywords like : Action,Adventure,Fantasy,Fiction
# List have multiple dictionaries
# this list is stored as string
# string to List
import ast
ast.literal_eval(movies.iloc[0,3])
```

```python
# function to get the keywords from all the dictonaries of each row
def convert(obj):
    Lt=[]
    for i in ast.literal_eval(obj):
        Lt.append(i['name'])
    return Lt
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.sample(3)
```

```python
movies.iloc[0,4]                      ## Keywords
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0,5]                  ## Cast
```

```python
def convert2(obj):
    Lt=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            Lt.append(i['name'])
            counter+=1
        else:
            break
    return Lt
```

```python
convert2(movies.iloc[0,5])
```

```python
movies['cast']=movies['cast'].apply(convert2)                       ## cast
```

```python
movies.head()
```

```python
def fetch_director(obj):
    Lt=[]
    counter=0
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            Lt.append(i['name'])
            break
    return Lt
```

```python
movies['crew']=movies['crew'].apply(fetch_director)                  ## director
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies_new=movies[['movie_id','title','tags']]
```

```python
movies_new['tags']=movies_new['tags'].apply(lambda x:" ".join(x))
```

## Text preprocessing

```python
movies_new['tags']=movies_new['tags'].apply(lambda x:x.lower())
```

```python
movies_new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_new['tags']=movies_new['tags'].apply(stem)
```

```python
movies_new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movies_new['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
movies_new['title']
```

```python
movie_index=movies_new[movies_new['title']=="Pirates of the Caribbean: At World's End"].index[0]
movie_index
```

```python
distances=similarity[1]
distances
```

```python
list(enumerate(distances))
```

```python
sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=movies_new[movies_new['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(movies_new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
